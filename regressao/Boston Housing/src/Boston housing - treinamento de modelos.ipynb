```python
# bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# carregando dados
df = pd.read_csv("../data/housing.csv", sep=',', encoding='utf-8')
```

```python
# check 
df.head(3)
```

```python
# check
df.shape
```

**previsores**

```python
# separação alvo x previsores
X_RM = df.iloc[:,0:1].values
X_LSTAT = df.iloc[:,1:2].values
y = df.iloc[:,3].values
```

```python
# valor medio dos imóveis
y
```

```python
# previsor em função do número de cômodos
X_RM
```

```python
# previsor em função da classe social
X_LSTAT
```

**Modelo em função do número de cômodos**

```python
# bases treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(X_RM, y, test_size = 0.3, random_state=10)
```

```python
# check shape
x_treino.shape, y_treino.shape
```

```python
# check shape
x_teste.shape, y_teste.shape
```

```python
# regressor simples
reg_linear = LinearRegression()
reg_linear.fit(x_treino, y_treino)
```

```python
# coeficiente linear
reg_linear.intercept_
```

```python
# coeficiente angular
reg_linear.coef_
```

```python
# coeficiente de determinação dados de treino 
# esse coeficiente explica o modelo
reg_linear.score(x_treino, y_treino)
```

```python
# coeficiente de determinação base de teste
reg_linear.score(x_teste, y_teste)
```

```python
# plot
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

titles = ['Treino', 'Teste']
x_data = [x_treino, x_teste]
y_data = [y_treino, y_teste]

for i in range(2):
    axs[i].scatter(y=y_data[i], x=x_data[i], color='blue', s=10, alpha=0.9)
    x_plot = np.linspace(4, 9)
    axs[i].plot(x_plot, x_plot*reg_linear.coef_ + reg_linear.intercept_, color='r')
    axs[i].set_title(titles[i])
    axs[i].set_ylabel('Valor médio')
    axs[i].set_xlabel('Quantidade média de cômodos')

plt.tight_layout()
plt.show()
```

```python
# metricas
previsoes_teste = reg_linear.predict(x_teste)
```

```python
# erro absoluto
mean_absolute_error(y_teste, previsoes_teste)
```

```python
# raiz do erro quadrático
np.sqrt(mean_squared_error(y_teste, previsoes_teste))
```

```python
# check modelo com valores distintos de cômodos
for i in range(3):
    comodos = np.random.randint(1,15)
    print(f'Comodos: {comodos} - Previsão: {reg_linear.predict([[comodos]])}')
```

**Modelo em função da classe social**

```python
# separando bases
x_treino, x_teste, y_treino, y_teste = train_test_split(X_LSTAT, y, test_size=0.3, random_state=10)
```

```python
# check
print(x_treino.shape, y_treino.shape)
print(x_teste.shape, y_teste.shape)
```

```python
# modelo
reg_linear_social = LinearRegression()
reg_linear_social.fit(x_treino, y_treino)
```

```python
# coef linear
reg_linear_social.intercept_
```

```python
# coef angular
reg_linear_social.coef_
```

```python
# coef determinacao treino
reg_linear_social.score(x_treino, y_treino)
```

```python
# coef determinicao teste
reg_linear_social.score(x_teste, y_teste)
```

```python
# metricas
previsao_teste = reg_linear_social.predict(x_teste)
```

```python
# erro abs
mean_absolute_error(y_teste, previsao_teste)
```

```python
# raiz erro quadrático
np.sqrt(mean_squared_error(y_teste, previsao_teste))
```

```python
# check modelo com valores distintos de classe social
for i in range(3):
    comodos = np.random.randint(20,55)
    print(f'Classe: {comodos} - Previsão: {reg_linear.predict([[comodos]])}')
```
